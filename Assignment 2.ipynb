# 1.Importing the Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('cars_data.csv')
```

```python
data
```

# The given data set contains:
     Data of purchase details of cars from 2009 to 2017 including its make,model,car gender,buyer gender,sale price,resell      price etc
     It contains a total 10000 rows x 20 columns

# 2.Checking the Nullvalues

```python
data.isnull().sum()
```

# Inference:
    There is no null values in the given data set

# 3.Bar graph of male vs female buyers participated in the sales

```python
p=sns.set_theme(style="darkgrid")
sns.countplot(x="Buyer Gender",hue='Buyer Gender',data=data)
plt.legend(loc='upper center')
plt.title('Male Vs Female Buyers Participated in the Sales',fontsize=16)
plt.xlabel('Buyer Gender',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks()
plt.yticks()
```

# Inference:
    1.Female buyers is more than Male Buyers
    2.The number of female buyers is more than 5000 while the number of male buyers is almost 5000

# 4.To Find the top 5 cars based on their sales price

```python
data1=data[["Make","Model","Sale Price"]]
data1.nlargest(5,"Sale Price")
```

# Inference
    1.The Companies having highest selling price are Volkswagen,Audi,Lexus and Mazda.
    2.Of these Lexus models hold the 3rd and 5th position.We can infer that Lexus models has a high selling price
    3.All models occupying the top 5 positions has a Selling Price above 99970

# 5.To Find the least 5 cars based on their Resell price.

```python
data2=data[["Make","Model","Sale Price","Resell Price"]]
data2.nsmallest(5,"Resell Price")
```

# Inference:
    1.The Brands having the least Resell Price are Chevrolet, GMC,Jaguar,Mitsubishi and Infinity.
    2.Only Jaguar XF has almost the same sale and resell price.It can be a good option for those who wants to buy a budget      friendly car with a profitable resale value.
    3.Other four brands have very high sale price with respect to its resell price.
