```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Data

```python
data=pd.read_excel('iris.xls')
```

```python
data.isna()
```

# Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

# 1.Histogram

```python
plt.figure(figsize=(9,9))
plt.hist(data['SL'],color='violet',rwidth=0.8)
plt.xlabel('SL',fontsize=9)
plt.ylabel('Frequency',fontsize=9)
plt.title('Distribution of SW',fontsize=16)
plt.xticks()
plt.yticks()
```

# 2.Scatter Plot

```python
plt.figure(figsize=(8,8))
plt.scatter(data['SW'],data['PL'],s=8,c='blue',marker='*')
plt.xlabel('SL',fontsize=10)
plt.ylabel('PL',fontsize=10)
plt.title('Distribution of SW Vs PL',fontsize=16)
plt.xticks()
plt.yticks()
```

# 3.Pair Plot

```python
sns.pairplot(data)
```

# 4.Scatter Plot

```python
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
plt.scatter(data['SL'],data['SW'],s=9,marker='*',c='violet')
plt.title('Scatter Plot SL Vs SW',fontsize=12)
plt.xlabel('SL',fontsize=12)
plt.ylabel('SW',fontsize=12)

plt.subplot(2,2,2)
plt.scatter(data['PL'],data['PW'],s=9,marker='^',c='red')
plt.title('Scatter Plot PL Vs PW',fontsize=12)
plt.xlabel('PL',fontsize=12)
plt.ylabel('PW',fontsize=12)

plt.subplot(2,2,3)
plt.scatter(data['SL'],data['PL'],s=9,marker='*',c='blue')
plt.title('Scatter Plot SL Vs PL',fontsize=12)
plt.xlabel('SL',fontsize=12)
plt.ylabel('PL',fontsize=12)

plt.subplot(2,2,4)
plt.scatter(data['PL'],data['PW'],s=9,marker='^',c='green')
plt.title('Scatter Plot PL Vs PW',fontsize=12)
plt.xlabel('PL',fontsize=12)
plt.ylabel('PW',fontsize=12)
plt.margins(0.4)
plt.subplots_adjust(left=0.9, right=2, bottom=1, top=2.5)
```

# 5.Heat Map

```python
sns.heatmap(data.corr(),annot=True)
plt.title('Heat Map',fontsize=14)
plt.xticks(rotation=45)
```

# Box Plot

```python
sns.boxplot(x='PL',y='PW',data=data)
plt.title('Box Plot PL Vs PW',fontsize=12)
plt.xlabel('PL',fontsize=12)
plt.ylabel('PW',fontsize=12)
plt.subplots_adjust(left=0.1, right=3, bottom=1, top=2.5)
```
